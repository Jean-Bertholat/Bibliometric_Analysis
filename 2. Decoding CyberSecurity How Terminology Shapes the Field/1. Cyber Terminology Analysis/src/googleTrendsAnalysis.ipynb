```python
import pandas as pd
import matplotlib.pyplot as plt

# Lecture du fichier CSV en ignorant la premire ligne qui contient la catgorie
# Remplacez 'data.csv' par le chemin vers votre fichier
df = pd.read_csv('../../dataset/requests_google/google trends.csv', skiprows=1)

# Conversion de la colonne 'Mois' en datetime pour une meilleure gestion des dates
df['Mois'] = pd.to_datetime(df['Mois'], format='%Y-%m')

# Calcul de l'cart entre 'cyber security' et 'cybersecurity'
df['Ecart'] = df['cyber security'] - df['cybersecurity']

# Affichage du DataFrame avec l'cart
print(df[['Mois', 'Ecart']])

# Trac de l'volution de l'cart dans le temps
plt.figure(figsize=(10, 6))
plt.plot(df['Mois'], df['Ecart'], marker='', linestyle='-')
plt.title("Evolution of the gap between 'cyber security' and 'cybersecurity'")
plt.xlabel('Years')
plt.ylabel("Gap")
plt.grid(True)
plt.show()
```
